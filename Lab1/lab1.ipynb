```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("watch_test.csv")
```

```python
dataset.drop(["uts"], axis = 1, inplace = True)
```

```python
# data cleaning
dataset.replace(0, dataset.mean(), inplace = True)
dataset.fillna(dataset.mean(), inplace = True)
dataset.drop_duplicates(inplace = True)
```

```python
#feature scaling
from sklearn import preprocessing
X = dataset.values
scaler = preprocessing.StandardScaler().fit(X)
#X_norm = preprocessing.scale(X)
X_norm = scaler.transform(X)
```

```python
#vizitilazation initalize
m = X_norm.shape[0]
n = X_norm.shape[1]
header = dataset.columns
```

```python
#plot
fig = plt.figure(figsize = (30,10))
for i in range(n):
    fig.add_subplot(2,round(n/2),i+1)
    plt.plot(np.arange(m), X_norm[:, i])
    plt.title(dataset.columns[i])
```

```python
#plot hist
fig = plt.figure(figsize = (30,10))
for i in range(n):
    fig.add_subplot(2,round(n/2),i+1)
    plt.hist(X_norm[:, i])
    plt.title(dataset.columns[i])
```

```python
# pairplot
import seaborn as sns; sns.set()
sns.pairplot(dataset, markers = 'o')
plt.show()
```
